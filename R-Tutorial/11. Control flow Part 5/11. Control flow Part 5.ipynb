We are almost there. Today, we will continue on control flow. So far, we already touched `for` loop,` while` loop, and `if-else`. This tutorial, we will talk about two little things, `break` and `next`.

These two statements I did not mention in loop because they need to use with `for` loop or `while` loop together. 

**Break**

`break` means break. In a loop, if we do not set any conditions, then the loop will iterate each value in the sequence, but, how could we stop loop early if some conditions are met? The answer is `break`. Here is a logical plot to show how `break` works.
<img src="r-break-flowchart.jpg"/>

```R
# For loop with break
for (item in sequence) {
  if (condition) {
    break
  }
  statement1
  statement2
  ...
}

# While loop with break
while (conditions) {
  if(condition) {
      break
  }
  count <- count + 1
  statement1
  statement2
  ...
}
```

So, the logic is simple.  In one loop, when `if` statement is true then break the current program which means the loop be terminated.  If not the `if`  statement is false, then `if` statement will do nothing to the whole loop.

**Next**

Sometime, in a loop, we would like to ignore some cases. In this case, `next` will help us.

```R
# for loop with next
for (item in sequence) {
  if (condition) {
      next
  }
  statement1
  statement2
  ...
}

# while loop with next
while (condition) {
  if (condition) {
      next
  }
  count <- count + 1
  statement1
  statement2
  ...
}
```

<img src="r-next-flowchart.png"/>

Here is the logic plot for the `next` statement.  In loop, when `if` statement is true, the loop will ignore this particular item, then move to next one. Here is different with break, the break will end loop but next will not.
